```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data (Replace this with your actual dataset)
# Assuming your dataset is named housing_data and has a column named 'year_built'
np.random.seed(123) # For reproducibility
housing_data = pd.DataFrame({'year_built': np.random.choice(range(1950, 2016), 500, replace=True)})

# Create decade bins
housing_data['decade'] = pd.cut(
    housing_data['year_built'], 
    bins=np.arange(1950, 2026, 10), 
    labels=[f"{i}s" for i in range(1950, 2016, 10)], 
    right=False
)

# Check for NA values in decade column due to improper binning
if housing_data['decade'].isna().any():
    print("Warning: Some values may not fall within the specified decade bins.")

# Count the number of houses per decade
year_counts = housing_data['decade'].value_counts().sort_index().reset_index()
year_counts.columns = ['Decade', 'Freq'] # Rename columns for clarity

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(year_counts['Decade'], year_counts['Freq'], color=plt.cm.rainbow(np.linspace(0, 1, len(year_counts))))
plt.title("Number of Houses Built per Decade (1950s - 2015)")
plt.xlabel("Decade")
plt.ylabel("Number of Houses Built")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (Replace this with your actual dataset)
merged_df = pd.DataFrame({'lotsize': [5000, 6000, 7000, 8000, 9000, 10000, 11000, 12000, 13000, 14000]})

# Create the box plot for "lotsize"
plt.figure(figsize=(8, 6))
plt.boxplot(merged_df["lotsize"].dropna(), vert=True, patch_artist=True, boxprops=dict(facecolor="skyblue"))

# Add labels and title
plt.title("Distribution of Lot Sizes", fontsize=16, fontweight="bold")
plt.ylabel("Lot Size (sq ft)", fontsize=14)
plt.xticks([1], ["Lot Size"], fontsize=12)

# Show the plot
plt.show()
```
