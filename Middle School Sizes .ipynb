```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the merged housing and schools dataset
housing_data = pd.read_csv("merged_housing_schools.csv")

# Convert school size and rating to numeric, handling any non-numeric values
housing_data['size_middle'] = pd.to_numeric(housing_data['size_middle'], errors='coerce')
housing_data['rating_middle'] = pd.to_numeric(housing_data['rating_middle'], errors='coerce')

# Create a scatter plot using seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=housing_data, x='size_middle', y='rating_middle', color='green', alpha=0.6)
plt.title("Scatter Plot of Middle School Size vs. Rating", fontsize=14)
plt.xlabel("Middle School Size", fontsize=12)
plt.ylabel("Middle School Rating", fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the merged housing and schools dataset
housing_data = pd.read_csv("merged_housing_schools.csv")

# Convert square footage to numeric, handling any non-numeric values
housing_data['sqft'] = pd.to_numeric(housing_data['sqft'], errors='coerce')

# Create a high-density plot using seaborn
plt.figure(figsize=(10, 6))
sns.kdeplot(data=housing_data, x='sqft', fill=True, color='blue', alpha=0.5)
plt.title("High Density Plot of Square Footage")
plt.xlabel("Square Footage (sq ft)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the merged housing and schools dataset
housing_data = pd.read_csv("merged_housing_schools.csv")

# Filter for condos only
condo_data = housing_data[housing_data['type'] == "condo"]

# Convert sold price to numeric, handling any non-numeric values
condo_data['soldprice'] = pd.to_numeric(condo_data['soldprice'], errors='coerce')

# Create a regression plot using seaborn
plt.figure(figsize=(10, 6))
sns.regplot(x='sqft', y='soldprice', data=condo_data, scatter_kws={'color': 'blue', 'alpha': 0.6}, line_kws={'color': 'red'})
plt.title("Regression Plot for Condo Prices and Square Footage")
plt.xlabel("Square Footage (sq ft)")
plt.ylabel("Sold Price (USD)")
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the merged housing and schools dataset
housing_data = pd.read_csv("merged_housing_schools.csv")

# Select relevant columns for clustering
cluster_data = housing_data[['neighborhood', 'high']]

# Convert categorical variables to factors (categorical type in pandas)
cluster_data['neighborhood'] = cluster_data['neighborhood'].astype('category')
cluster_data['high'] = cluster_data['high'].astype('category')

# Visualize using a cluster plot
plt.figure(figsize=(12, 8))
sns.stripplot(
    data=cluster_data,
    x='neighborhood',
    y='high',
    jitter=True, # Mimics geom_jitter
    alpha=0.7,
    palette="Set2" # Optional color palette
)

# Customize plot aesthetics
plt.title("Cluster Plot of Neighborhoods and High Schools")
plt.xlabel("Neighborhood")
plt.ylabel("High School")
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```
